```python
import pandas as pd
df = pd.read_csv('data.csv', sep=',')
print(df.values)
```

```python
# printing data frame
print("Data frame")
print(df)
```

```python
# Get unique values of a column
print("Unique values of a column")
uniqueValues = df['itemname'].unique()
print(uniqueValues)
```

```python
# Create dataframe with 'monitordate' and all the values in uniqueValues as header
print("Create dataframe with 'monitordate' and all the values in uniqueValues as header")
df2 = pd.DataFrame(columns = ['monitordate'])
for i in uniqueValues:
    df2[i] = 0
print(df2)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Go through the original data frame and fill the new data frame
print("Go through the original data frame and fill the new data frame")
for index, row in df.iterrows():
    # print(row['monitordate'], row['itemname'], row['concentration'])
    # if monitordate not found in df2, add it
    if not df2['monitordate'].str.contains(row['monitordate']).any():
        # get df2 column count and create an empty array with 0
        empty_row = [0] * len(df2.columns)
        # add monitordate to the first element of the array
        empty_row[0] = row['monitordate']
        # add the array to the end of the dataframe
        df2.loc[len(df2)] = empty_row
    df2.loc[df2['monitordate'] == row['monitordate'], row['itemname']] = row['concentration']
print(df2)
```
